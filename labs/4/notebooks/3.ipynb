```python
def get_first_letter_cabin(cabin):
    return cabin[0] if not pd.isna(cabin) else None
```

```python
train_df["cabin_first_letter"] = train_df["cabin"].apply(get_first_letter_cabin)
test_df["cabin_first_letter"] = test_df["cabin"].apply(get_first_letter_cabin)
```

```python
def get_first_letters_ticket(ticket):
    return ticket.split()[0] if ticket.split()[0][0].isalpha() else None
```

```python
train_df["ticket_first_letters"] = train_df["ticket"].apply(get_first_letters_ticket)
test_df["ticket_first_letters"] = test_df["ticket"].apply(get_first_letters_ticket)
```

```python
train_df["ticket_first_letters"].value_counts()
```

```python
from functools import partial
```

```python
def get_first_digits_ticket(ticket, digits):
    return (
        str(ticket.split()[-1])[:digits]
        if len(ticket.split()) > 1
        else str(ticket)[:digits]
    )
```

```python
get_first_digit_ticket = partial(get_first_digits_ticket, digits=1)
get_first_digits_ticket = partial(get_first_digits_ticket, digits=2)
```

```python
train_df["first_digit_ticket"] = train_df["ticket"].apply(get_first_digit_ticket)
test_df["first_digit_ticket"] = test_df["ticket"].apply(get_first_digit_ticket)
train_df["first_digits_ticket"] = train_df["ticket"].apply(get_first_digits_ticket)
test_df["first_digits_ticket"] = test_df["ticket"].apply(get_first_digits_ticket)
```

```python
useless_columns = ["name", "ticket", "cabin"]
train_df = train_df.drop(useless_columns, axis=1)
test_df = test_df.drop(useless_columns, axis=1)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
categorical_features = [
    "sex",
    "pclass",
    "embarked",
]  # 'cabin_first_letter', 'ticket_first_letters', 'first_digits_ticket', 'first_digit_ticket']
numeric_features = ["age", "sibsp", "parch", "fare"]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            Pipeline(
                steps=[
                    ("imputer", SimpleImputer(strategy="median")),
                    ("scaler", StandardScaler()),
                ]
            ),
            numeric_features,
        ),
        (
            "cat",
            Pipeline(
                steps=[
                    ("imputer", SimpleImputer(strategy="most_frequent")),
                    ("encoder", OneHotEncoder(handle_unknown="ignore")),
                ]
            ),
            categorical_features,
        ),
    ]
)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
```

```python
base_models = [
    ("rf", RandomForestClassifier(n_estimators=100, random_state=42)),
    ("gb", GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ("svm", SVC(kernel="rbf", C=1.0, probability=True, random_state=42)),
]

meta_model = LogisticRegression()

stacking_model = StackingClassifier(
    estimators=base_models, final_estimator=meta_model, cv=10
)

pipeline = Pipeline(
    steps=[("preprocessor", preprocessor), ("classifier", stacking_model)]
)
```

```python
pipeline
```

```python
%%time

pipeline.fit(train_df.drop("survived", axis=1), train_df["survived"])
y_pred = pipeline.predict(test_df.drop("survived", axis=1))
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(test_df["survived"], y_pred)
print(report)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class RareCategoryEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0.05):
        self.threshold = threshold
        self.replacement = "Other"

    def fit(self, X, y=None):
        self.columns_ = np.arange(X.shape[1])

        self.categories_ = {}
        for col in self.columns_:
            counts = pd.Series(X[:, col]).value_counts(normalize=True)

            rare_categories = counts[counts < self.threshold].index
            self.categories_[col] = rare_categories
        return self

    def transform(self, X):
        X_transformed = X.copy()
        for col in self.columns_:
            X_transformed[:, col] = np.where(
                np.isin(X_transformed[:, col], self.categories_[col]),
                self.replacement,
                X_transformed[:, col],
            )
        return X_transformed
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            Pipeline(
                steps=[
                    ("imputer", SimpleImputer(strategy="median")),
                    ("scaler", StandardScaler()),
                ]
            ),
            numeric_features,
        ),
        (
            "cat",
            Pipeline(
                steps=[
                    ("imputer", SimpleImputer(strategy="most_frequent")),
                    ("rare", RareCategoryEncoder(threshold=0.05)),
                    ("encoder", OneHotEncoder(handle_unknown="ignore")),
                ]
            ),
            categorical_features,
        ),
    ]
)
```

```python
base_models = [
    ("rf", RandomForestClassifier(n_estimators=100, random_state=42)),
    ("gb", GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ("svm", SVC(kernel="rbf", C=1.0, probability=True, random_state=42)),
]

meta_model = LogisticRegression()

stacking_model = StackingClassifier(
    estimators=base_models, final_estimator=meta_model, cv=10
)

pipeline = Pipeline(
    steps=[("preprocessor", preprocessor), ("classifier", stacking_model)]
)
```

```python
%%time

pipeline.fit(train_df.drop("survived", axis=1), train_df["survived"])
y_pred = pipeline.predict(test_df.drop("survived", axis=1))
```

```python
report = classification_report(test_df["survived"], y_pred)
print(report)
```
