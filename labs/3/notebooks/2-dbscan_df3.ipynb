```python
import pandas as pd
```

```python
df_3 = pd.read_csv("../data/clustering/clustering_3.csv", sep="\t", header=None)
df_3
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalled_df_3 = StandardScaler().fit_transform(df_3)
scalled_df_3
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
cluster = DBSCAN(eps=0.25).fit_predict(scalled_df_3)
plt.scatter(x=scalled_df_3[:, 0], y=scalled_df_3[:, 1], c=cluster)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
silhouette_score(scalled_df_3, cluster), davies_bouldin_score(scalled_df_3, cluster)
```
