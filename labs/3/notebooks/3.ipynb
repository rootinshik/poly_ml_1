```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
```

```python
image = cv2.imread("../data/th-504933578.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
pixels = image.reshape(-1, 3).astype(np.float32)
```

```python
%%time

n_clusters = 8
kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(pixels)

compressed_pixels = kmeans.cluster_centers_[labels].astype("uint8")
compressed_image = compressed_pixels.reshape(image.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 10))
ax[0].imshow(image)
ax[0].set_title(" ", fontsize=14)
ax[0].axis("off")

ax[1].imshow(compressed_image)
ax[1].set_title(f"  ({n_clusters} )", fontsize=14)
ax[1].axis("off")

plt.tight_layout()
plt.show();
```
