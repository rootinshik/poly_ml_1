```python
import pandas as pd
```

```python
df = pd.read_csv("../data/votes.csv")
```

```python
df
```

```python
df = df.fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_scaled = StandardScaler().fit_transform(df)
```

```python
df_scaled
```

```python
from scipy.cluster.hierarchy import linkage
```

```python
Z = linkage(df_scaled, method="ward", metric="euclidean")
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize=(15, 10))
dendrogram(Z, labels=df.index, leaf_rotation=90, leaf_font_size=8)
plt.title("Дендрограмма голосования за республиканцев (1856–1976)")
plt.xlabel("Штаты")
plt.ylabel("Расстояние между кластерами")
plt.show()
```
