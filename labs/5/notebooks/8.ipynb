```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cars.csv")
df.head(3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["speed"], df["dist"])
plt.xlabel("Скорость (мили/час)")
plt.ylabel("Дистанция торможения (футы)")
plt.title("Зависимость дистанции торможения от начальной скорости")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(df[["speed"]].values, df["dist"].values)
```

```python
import numpy as np
```

```python
speed_40 = np.array([[40]])
model.predict(speed_40)
```
